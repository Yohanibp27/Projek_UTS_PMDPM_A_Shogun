### 3. Data Preparation

Hanlde missing value

```python
numeric_cols = df_dataUTS2.select_dtypes(include=[np.number]).columns
categorical_cols = df_dataUTS2.select_dtypes(exclude=[np.number]).columns

# data Null
df_dataUTS2[numeric_cols] = df_dataUTS2[numeric_cols].fillna(df_dataUTS[numeric_cols].median())
for col in categorical_cols:
    df_dataUTS2[col] = df_dataUTS2[col].fillna(df_dataUTS2[col].mode()[0])

# data kosong
df_dataUTS2.dropna(subset=['basement', 'attic'], inplace=True)
# data NaN
df_dataUTS2.fillna(method='ffill', inplace=True)
# Memastikan tidak ada nilai NaN yang tersisa setelah penanganan
print("\nJumlah NaN setelah penanganan:\n", df_dataUTS2.isnull().sum())
```

Menghapus data outlier

```python
from pandas.api.types import is_numeric_dtype

def remove_outlier(df_dataUTS2):
    for col_name in list(df_dataUTS2.columns):
        if is_numeric_dtype(df_dataUTS2[col_name]):
            Q1 = df_dataUTS2[col_name].quantile(0.25)
            Q3 = df_dataUTS2[col_name].quantile(0.75)

            IQR = Q3 - Q1
            batas_atas = Q1 - 1.5 * IQR
            batas_bawah = Q3 + 1.5 * IQR
            df_dataUTS2 = df_dataUTS2.loc[(df_dataUTS2[col_name] > batas_bawah) & (df_dataUTS2[col_name] < batas_atas)]
    return df_dataUTS2

df_dataUTS_clean = remove_outlier(df_dataUTS2)
print("Jumlah baris DataFrame sebelum dibuat outlier", df_dataUTS2.shape[0])
print("Jumlah baris DataFrame setelah dibuat outlier", df_dataUTS_clean.shape[0])

# membalik sumbu y
plt.gca().invert_yaxis()
plt.show()
```

### 4. Data Encoding

Encoding kolom kategori menggunakan teknik One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.feature_selection import SelectKBest, chi2

# kolom kategori
cat_cols = ["hasyard", "haspool", "isnewbuilt", "hasstormprotector", "hasstorageroom"]
# init OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')
column_transformer = make_column_transformer(
    (encoder, cat_cols),
    remainder='passthrough'
)
# transformasi data latih dan uji
X_classif_train_enc = column_transformer.fit_transform(X_classif_train)
X_classif_test_enc = column_transformer.transform(X_classif_test)

df_train_enc = pd.DataFrame(X_classif_train_enc, columns=column_transformer.get_feature_names_out())
df_test_enc = pd.DataFrame(X_classif_test_enc, columns=column_transformer.get_feature_names_out())

df_train_enc.head(10)
df_test_enc.head(10)
```

Memilih fitur terbaik dari dataset yang telah di-encode dan mempersiapkan data untuk pelatihan model

```python
encoded_feature_names = column_transformer.get_feature_names_out()

# dataFrame data latih dan uji
X_classif_train_enc_df = pd.DataFrame(X_classif_train_enc, columns=encoded_feature_names)
X_classif_test_enc_df = pd.DataFrame(X_classif_test_enc, columns=encoded_feature_names)

X_classif_train_enc_df['label'] = y_classif_train.values
X_classif_test_enc_df['label'] = y_classif_test.values

# fitur terbaik menggunakan SelectKBest
X = X_classif_train_enc_df.drop("label", axis=1)
y = X_classif_train_enc_df["label"]

# Memilih k fitur terbaik
selector = SelectKBest(score_func=chi2, k=5)
X_new = selector.fit_transform(X, y)

# Menampilkan fitur yang dipilih
selected_features = X.columns[selector.get_support()]
print("Fitur yang dipilih:")
print(selected_features)

# Menyiapkan data untuk model
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.25, random_state=78)
```

### 5. Model Pepiline
### Notebook 2: Gradient Boosting dan Support Vector Machine
Model Klasifikasi
- Data scaling: StandarScaler dan MinMaxScaler
- Feature selection: SelecBest dan SelectPercentile 

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.feature_selection import SelectKBest, SelectPercentile, f_classif

pipe_GBT = Pipeline([
    ('feat_select', SelectKBest(score_func=f_classif)),
    ('clf', GradientBoostingClassifier(random_state=78))
])

params_grid_GBT = [
    {
        'feat_select': [SelectKBest(score_func=f_classif)],
        'feat_select__k': np.arange(2, 6), 
        'clf__max_depth': [3, 4],  
        'clf__n_estimators': [100],
        'clf__learning_rate': [0.01, 0.1]
    },
    {
        'feat_select': [SelectPercentile(score_func=f_classif)],
        'feat_select__percentile': np.arange(20, 50, 10),
        'clf__max_depth': [3, 4],
        'clf__n_estimators': [100],
        'clf__learning_rate': [0.01, 0.1]
    }
]

GSCV_GBT = GridSearchCV(pipe_GBT, params_grid_GBT, cv=StratifiedKFold(n_splits=5))
GSCV_GBT.fit(X_classif_train_enc, y_classif_train)
print("GSCV GBT finished")
```

Support Vector Classifier 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectPercentile, SelectKBest
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline

pipeline_SVM = Pipeline([
    ('scale', MinMaxScaler()),
    ('feat_select', SelectKBest()),
    ('clf', SVC(class_weight='balanced'))
])

params_grid_SVM = [
    {
        'scale': [MinMaxScaler()],
        'feat_select__k': np.arange(2, 6),
        'clf__kernel': ['rbf'],
        'clf__C': [0.1, 1],
        'clf__gamma': [0.1]
    },
    {
        'scale': [StandardScaler()],
        'feat_select': [SelectPercentile()],
        'feat_select__percentile': [30], 
        'clf__kernel': ['rbf'],
        'clf__C': [0.1, 1],
        'clf__gamma': [0.1, 1]
    }
]

SKF = StratifiedKFold(n_splits=5, shuffle=True, random_state=78)

GSCV_SVM = GridSearchCV(pipeline_SVM, params_grid_SVM, cv=SKF)
GSCV_SVM.fit(X_classif_train_enc, y_classif_train)
print("GSCV SVM finished")
```

### 6. Model Training and Evaluation
### Notebook 2: Gradient Boosting dan Support Vector Machine
Model Klasifikasi
- Confusion matrix
- Metrik akurasi
- Precision
- Recall
- F1-score

Gradient Boosting

```python
from sklearn.metrics import classification_report

# metrik evaluasi
print("CV Score: {}".format(GSCV_GBT.best_score_))
print("Test Score: {}".format(GSCV_GBT.best_estimator_.score(X_classif_test_enc, y_classif_test)))
print("Best model:", GSCV_GBT.best_estimator_)

# fitur terbaik
mask = GSCV_GBT.best_estimator_.named_steps['feat_select'].get_support()
print("Best features:", df_train_enc.columns[mask])

# prediksi dan evaluasi
GBT_pred = GSCV_GBT.predict(X_classif_test_enc)

# precision, recall, F1-score, accuracy
print("Confusion matrix GBT: \n", y_classif_test, GBT_pred)
print("Classification report GBT: \n", classification_report(y_classif_test, GBT_pred))
```

Gradient Boosting

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_classif_test, GBT_pred, labels=GSCV_GBT.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=GSCV_GBT.classes_)
disp.plot()

plt.title("Gradient Boosting Confusion Matrix")
plt.show()

# precision, recall, dan F1-score
print("Classification Report GBT: \n", classification_report(y_classif_test, GBT_pred))
```

```python
# metrik evaluasi
print("CV Score: {}".format(GSCV_SVM.best_score_))
print("Test Score: {}".format(GSCV_SVM.best_estimator_.score(X_classif_test_enc, y_classif_test)))
print("Best model: {}", GSCV_SVM.best_estimator_)

# fitur terbaik
mask = GSCV_SVM.best_estimator_.named_steps['feat_select'].get_support()
print("Best features", df_train_enc.columns[mask])

# prediksi dan evaluasi
SVM_pred = GSCV_SVM.predict(X_classif_test_enc)

# precision, recall, F1-score, accuracy
print("Confusion matrix SVM: \n", y_classif_test, SVM_pred)
print("Classification report SVM: \n", classification_report(y_classif_test, SVM_pred))
```
